```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.callbacks import EarlyStopping

# Gerar dados aleatórios para recursos e rótulos
num_pontos = 100
features = np.linspace(0, 10, num_pontos)
labels = 3 * features + 2 + np.random.normal(0, 0.5, num_pontos)  # Rótulos com ruído

# Dividir os dados em conjuntos de treinamento, validação e teste
num_treinamento = int(0.7 * num_pontos)
num_validacao = int(0.15 * num_pontos)
num_teste = num_pontos - num_treinamento - num_validacao

dados_treinamento = (features[:num_treinamento], labels[:num_treinamento])
dados_validacao = (features[num_treinamento:num_treinamento + num_validacao], labels[num_treinamento:num_treinamento + num_validacao])
dados_teste = (features[num_treinamento + num_validacao:], labels[num_treinamento + num_validacao:])

# Criar o modelo da rede neural com regularização L2
modelo_com_regularizacao = tf.keras.Sequential([
    tf.keras.layers.Dense(units=16, input_shape=[1], activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.0001)),
    tf.keras.layers.Dense(units=8, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
    #tf.keras.layers.Dense(units=4, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
    tf.keras.layers.Dense(units=1)
])

# Compilar o modelo com uma taxa de aprendizado menor e otimizador separado
otimizador_com_regularizacao = tf.keras.optimizers.Adam(learning_rate=0.01)
modelo_com_regularizacao.compile(optimizer=otimizador_com_regularizacao, loss='mean_squared_error')

# Definir o Early Stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Treinar o modelo com um número maior de épocas
historico_com_regularizacao = modelo_com_regularizacao.fit(dados_treinamento[0], dados_treinamento[1], epochs=2000,
                                                           validation_data=dados_validacao, callbacks=[early_stopping])

# Criar o modelo da rede neural sem regularização
modelo_sem_regularizacao = tf.keras.Sequential([
    tf.keras.layers.Dense(units=16, input_shape=[1], activation='relu'),
    tf.keras.layers.Dense(units=8, activation='relu'),
    #tf.keras.layers.Dense(units=4, activation='relu'),
    tf.keras.layers.Dense(units=1)
])

# Compilar o modelo sem regularização e otimizador separado
otimizador_sem_regularizacao = tf.keras.optimizers.Adam(learning_rate=0.01)
modelo_sem_regularizacao.compile(optimizer=otimizador_sem_regularizacao, loss='mean_squared_error')

# Treinar o modelo sem regularização
historico_sem_regularizacao = modelo_sem_regularizacao.fit(dados_treinamento[0], dados_treinamento[1], epochs=2000,
                                                           validation_data=dados_validacao, callbacks=[early_stopping])

# Avaliar ambos os modelos no conjunto de teste
perda_teste_com_regularizacao = modelo_com_regularizacao.evaluate(dados_teste[0], dados_teste[1])
perda_teste_sem_regularizacao = modelo_sem_regularizacao.evaluate(dados_teste[0], dados_teste[1])

# Plotar os resultados com e sem regularização
plt.figure(figsize=(10, 6))
plt.scatter(features, labels, label='Dados Originais')

# Plotar previsões com regularização em vermelho
plt.plot(features, modelo_com_regularizacao.predict(features), color='red', label=f'Predições com Regularização (Erro de Teste: {perda_teste_com_regularizacao:.4f})')

# Plotar previsões sem regularização em verde
plt.plot(features, modelo_sem_regularizacao.predict(features), color='green', label=f'Predições sem Regularização (Erro de Teste: {perda_teste_sem_regularizacao:.4f})')

plt.title('Regressão usando Rede Neural com e sem Regularização')
plt.xlabel('Features')
plt.ylabel('Labels')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(historico_com_regularizacao.history['loss'], label='Treinamento com Regularização')
plt.plot(historico_sem_regularizacao.history['loss'], label='Treinamento sem Regularização')
plt.title('Evolução nos Erros de Treinamento')
plt.xlabel('Época')
plt.ylabel('Erro de Treinamento')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.callbacks import EarlyStopping

# Gerar dados aleatórios para recursos e rótulos
num_pontos = 100
features = np.linspace(0, 10, num_pontos)
labels = (features > 5).astype(int)  # Classificação binária: 1 se > 5, 0 se <= 5

# Dividir os dados em conjuntos de treinamento, validação e teste
num_treinamento = int(0.7 * num_pontos)
num_validacao = int(0.15 * num_pontos)

# Separação equilibrada das classes para teste, validação e treinamento
indices_classe_0 = np.where(labels == 0)[0]
indices_classe_1 = np.where(labels == 1)[0]

# Dividir os índices de cada classe para os conjuntos
num_indices_val = int(0.15 * len(indices_classe_0))  # Número de índices para validação
num_indices_teste = int(0.15 * len(indices_classe_1))  # Número de índices para teste
num_indices_treinamento = len(labels) - num_indices_val - num_indices_teste  # Restante para treinamento

# Embaralhar os índices antes de dividir
np.random.shuffle(indices_classe_0)
np.random.shuffle(indices_classe_1)

# Dividir os índices para os conjuntos
indices_val = np.concatenate((indices_classe_0[:num_indices_val], indices_classe_1[:num_indices_val]))
indices_teste = np.concatenate((indices_classe_0[num_indices_val:num_indices_val + num_indices_teste], indices_classe_1[num_indices_val:num_indices_val + num_indices_teste]))
indices_treinamento = np.concatenate((indices_classe_0[num_indices_val + num_indices_teste:], indices_classe_1[num_indices_val + num_indices_teste:]))

# Embaralhar novamente os índices para evitar ordens específicas
np.random.shuffle(indices_val)
np.random.shuffle(indices_teste)
np.random.shuffle(indices_treinamento)

# Usar os índices para obter os conjuntos de dados
dados_validacao= (features[indices_val], labels[indices_val])
dados_teste = (features[indices_teste], labels[indices_teste])
dados_treinamento = (features[indices_treinamento], labels[indices_treinamento])


# Criar o modelo da rede neural com regularização L2
modelo_com_regularizacao = tf.keras.Sequential([
    tf.keras.layers.Dense(units=16, input_shape=[1], activation='sigmoid', kernel_regularizer=tf.keras.regularizers.l2(0.0001)),
    tf.keras.layers.Dense(units=8, activation='sigmoid', kernel_regularizer=tf.keras.regularizers.l2(0.0001)),
    tf.keras.layers.Dense(units=8, activation='sigmoid', kernel_regularizer=tf.keras.regularizers.l2(0.0001)),
    tf.keras.layers.Dense(units=4, activation='sigmoid', kernel_regularizer=tf.keras.regularizers.l2(0.0001)),
    tf.keras.layers.Dense(units=1, activation='sigmoid')  # Saída com ativação sigmoid para classificação binária
])

# Compilar o modelo com uma taxa de aprendizado menor e otimizador separado
otimizador_com_regularizacao = tf.keras.optimizers.Adam(learning_rate=0.001)
modelo_com_regularizacao.compile(optimizer=otimizador_com_regularizacao, loss='binary_crossentropy', metrics=['accuracy'])

# Definir o Early Stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Treinar o modelo com um número maior de épocas
historico_com_regularizacao = modelo_com_regularizacao.fit(dados_treinamento[0], dados_treinamento[1], epochs=2000,
                                                           validation_data=dados_validacao, callbacks=[early_stopping])

# Criar o modelo da rede neural sem regularização
modelo_sem_regularizacao = tf.keras.Sequential([
    tf.keras.layers.Dense(units=16, input_shape=[1], activation='sigmoid'),
    tf.keras.layers.Dense(units=8, activation='sigmoid'),
    tf.keras.layers.Dense(units=8, activation='sigmoid'),
    tf.keras.layers.Dense(units=4, activation='sigmoid'),
    tf.keras.layers.Dense(units=1, activation='sigmoid')  # Saída com ativação sigmoid para classificação binária
])

# Compilar o modelo sem regularização e otimizador separado
otimizador_sem_regularizacao = tf.keras.optimizers.Adam(learning_rate=0.001)
modelo_sem_regularizacao.compile(optimizer=otimizador_sem_regularizacao, loss='binary_crossentropy', metrics=['accuracy'])

# Treinar o modelo sem regularização
historico_sem_regularizacao = modelo_sem_regularizacao.fit(dados_treinamento[0], dados_treinamento[1], epochs=2000,
                                                           validation_data=dados_validacao, callbacks=[early_stopping])

# Avaliar ambos os modelos no conjunto de teste
perda_teste_com_regularizacao, acuracia_teste_com_regularizacao = modelo_com_regularizacao.evaluate(dados_teste[0], dados_teste[1])
perda_teste_sem_regularizacao, acuracia_teste_sem_regularizacao = modelo_sem_regularizacao.evaluate(dados_teste[0], dados_teste[1])

# Plotar os resultados com e sem regularização
plt.figure(figsize=(10, 6))
plt.scatter(features, labels, label='Dados Originais')

# Plotar previsões com regularização em vermelho
plt.plot(features, modelo_com_regularizacao.predict(features), color='red', label=f'Predições com Regularização (Acurácia de Teste: {acuracia_teste_com_regularizacao:.4f})')

# Plotar previsões sem regularização em verde
plt.plot(features, modelo_sem_regularizacao.predict(features), color='green', label=f'Predições sem Regularização (Acurácia de Teste: {acuracia_teste_sem_regularizacao:.4f})')

plt.title('Classificação usando Rede Neural com e sem Regularização')
plt.xlabel('Features')
plt.ylabel('Labels')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(historico_com_regularizacao.history['accuracy'], label='Treinamento com Regularização')
plt.plot(historico_sem_regularizacao.history['accuracy'], label='Treinamento sem Regularização')
plt.title('Evolução nos Erros de Treinamento')
plt.xlabel('Época')
plt.ylabel('Erro de Treinamento')
plt.legend()
plt.grid(True)
plt.show()
```
